```python
import pandas as pd
import numpy as np
```

# Load Data

```python
file_path = 'your file path'
df = pd.read_excel(file_path)
```

```python
#using field that can be calculated
df[['Order ID','Sales', 'Quantity', 'Profit']].head()
```

# Summation of 'Sales' and 'Profit'

```python
#1. Basic Artithmetic
df['Total_Revenue'] = df['Sales'] + df['Profit']
```

```python
#2. Using 'eval' method
df.eval('Total_Revenue = Sales + Profit', inplace=True)
```

```python
#3. Using 'assign' method
df = df.assign(Total_Revenue=df['Sales'] + df['Profit'])
```

```python
df[['Order ID','Sales', 'Quantity', 'Profit','Total_Revenue']].head()
```

# Squares and Square Roots of 'Sales'

```python
#4. Use a Function to Square Sales
def square_value(x):
    return x ** 2

df['squared_sales'] = df['Sales'].apply(square_value)
```

```python
#5. Use a Numpy Function to Square Roots Sales
import numpy as np

df['sqrt_sales'] = np.sqrt(df['Sales'])
```

```python
df[['Order ID','Sales', 'squared_sales','sqrt_sales']].head()
```
